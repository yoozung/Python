```python
#버스번호로 정보추출 함수

url = 'http://ws.bus.go.kr/api/rest/' #기본 url/명령어?params

def makeUrl(cmd, params):#수행하고 싶은 기능에 맞게 url 변경해줌
    u = url
    u += cmd
    for c in params:
        u += c
    return u

def printPath(bus_id):
    cmd = 'busRouteInfo/getStaionByRoute'
    params = ['?ServiceKey='+api_key, '&busRouteId='+bus_id]
    u = makeUrl(cmd, params)
    html = requests.get(u).text #url에 요청
    root = BeautifulSoup(html, 'lxml-xml')
    code = root.find('headerCd').get_text()
    if code=='0':
        items = root.find_all('itemList')#배열
        for item in items:
            seq = item.find('seq').get_text()
            stationName = item.find('stationNm').get_text()
            print(seq, '.', stationName)
    else:
        print('오류발생 code:', code)
        
def getInfoByBusNum(num): #버스 번호를 파람으로 전달하면 그 버스 정보를 출력
    cmd = 'busRouteInfo/getBusRouteList'
    params = ['?ServiceKey='+api_key, '&strSrch='+num]
    u = makeUrl(cmd, params)
    html = requests.get(u).text #url에 요청
    root = BeautifulSoup(html, 'lxml-xml')
    code = root.find('headerCd').get_text()
    if code=='0':
        bus_id = root.find('busRouteId').get_text()
        bus_name = root.find('busRouteNm').get_text()
        edstation = root.find('edStationNm').get_text()
        ststation = root.find('stStationNm').get_text()
        firsttm= root.find('firstBusTm').get_text()
        lasttm= root.find('lastBusTm').get_text()
        term = root.find('term').get_text()
        
        print('노선명:', bus_name)
        print('배차간격:', term)
        print('기점:', ststation)
        print('종점:', edstation)
        print('찻차시간:', firsttm)
        print('막차시간:', lasttm)
        print('====경유지====')
        printPath(bus_id)

    else:
        print('오류발생 code:', code)
```

```python
def getArsIdByStName(st_name):
    cmd = 'stationinfo/getStationByName'
    params = ['?ServiceKey='+api_key, '&stSrch='+st_name]
    u = makeUrl(cmd, params)
    html = requests.get(u).text #url에 요청
    root = BeautifulSoup(html, 'lxml-xml')
    code = root.find('headerCd').get_text()
    st_arsid = []
    if code=='0':
        items = root.find_all('itemList')#배열
        for item in items:
            stNm = item.find('stNm').get_text()
            if st_name == stNm:
                id = item.find('arsId').get_text()
                st_arsid.append(id)
    else:
        print('error code:', code)
    return st_arsid
```

```python
html = getArsIdByStName('용산역')
html
```

```python
def getBusListByStArsid(arsid):
    cmd = 'stationinfo/getRouteByStation'
    params = ['?ServiceKey='+api_key, '&arsId='+arsid]
    u = makeUrl(cmd, params)
    html = requests.get(u).text #url에 요청
    root = BeautifulSoup(html, 'lxml-xml')
    code = root.find('headerCd').get_text()
    bus = []
    if code=='0':
        items = root.find_all('itemList')#배열
        for item in items:
            busNm = item.find('busRouteNm').get_text()
            bus.append(busNm)
        return bus
    else:
        print('error code:', code)
```

```python
html = getBusListByStArsid('03255')
html
```

```python
arsid = getArsIdByStName('용산역')
for a in arsid:
    busList = getBusListByStArsid(a)
    print(a,' 정거장에 정거하는 버스 목록')
    for b in busList:
        print(b)
```
