1. csv 다운파일 분석

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#csv파일의 내용을 읽어서 데이터프레임으로 변환
df = pd.read_csv('aed.csv', encoding='euc-kr', index_col='NO')
df
```

```python
def getByPhoneNumber(number):
    df = pd.read_csv('aed.csv', encoding='euc-kr', index_col='NO')
    row = df[df['전화번호']==number]
    if row.empty :
        print('없는 전화 번호')
    else:
        print(row)
```

```python
getByPhoneNumber('02-762-8485')
```

```python
getByPhoneNumber('134324')
```

2. excel 다운 파일 분석

```python
df = pd.read_excel('서울교통현황.xlsx', engine='openpyxl') 
df[:30]
```

```python
y = df.pivot_table(index=['년월', '노선명'])
y
```

```python
x = df.pivot(index='년월', columns='노선명', values='승차총승객수')
x
```
