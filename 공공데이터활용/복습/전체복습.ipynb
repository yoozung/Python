```python
def getByPhoneNumber(number):
    df = pd.read_csv('aed.csv', encoding='euc-kr', index_col='NO')
    row = df[df['전화번호']==number]
    if row.empty :
        print('없는 전화 번호')
    else:
        print(row)
```
